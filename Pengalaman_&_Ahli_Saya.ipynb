```python

import pandas as pd
import numpy as np

dates = ["2011", "2013", "2014", "2016", "2017", "2019", "2021",
         "2022", "2024"]

phones = ["Rakit PC, Install Config Windows, Linux, Jaringan", "Ujicoba Pembuatan Virus & Hacking tojan dari linux", "Penetrasi Trojan & Ujicoba Exploit, Servis PC & Instalasi Windows", "Jualan CD Software Komputer(Keliling & Kakilima)", "Pengeditan Foto & Disain Grafis, Project FrutyLoop Musik Remix", "Buka Warnet, Disain Grafis, Minning Crypto, Upgrade & Update Website Perikanan", "Disain Website & Aplikasi, Disain Aplikasi Web Apps Fish Store",
          "Disain Aplikasi android & IOS, Pembuatan Landing Page dan update data Perusahaan, Penelitian Data, BUMN Forum Digital Talent, Pengurus Kontrakan, Periset Data Online BRIN", "Peneliti Data BRIN yang telah di legalitas oleh BUMN, Periset & Pengembang Sistem Pakar A.I."]

iphone_df = pd.DataFrame(data={"Date": dates, "Pengalaman": phones})
iphone_df["Date"] = pd.to_datetime(iphone_df["Date"])
iphone_df["Stabil"] = [np.random.randint(-90,-50) if (i%2)==0 else np.random.randint(50,100) for i in range(len(iphone_df))]

iphone_df
```

```python
import matplotlib.pyplot as plt

with plt.style.context("fivethirtyeight"):
    fig, ax = plt.subplots(figsize=(18,9))

    ax.plot(iphone_df.Date, [0,]* len(iphone_df), "-o", color="black", markerfacecolor="white");

    ax.set_xticks(pd.date_range("2011-01-01", "2023-01-01", freq="ys"), range(2011, 2024));
    ax.set_ylim(-7,7);

    for idx in range(len(iphone_df)):
        dt, Pengalaman, Stabil = iphone_df["Date"][idx], iphone_df["Pengalaman"][idx], iphone_df["Stabil"][idx]
        dt_str = dt.strftime("%b-%Y")
        ax.annotate(dt_str + "\n" + Pengalaman, xy=(dt, 0.1 if Stabil>0 else -0.1),xytext=(dt, Stabil),
                    arrowprops=dict(arrowstyle="-",color="red", linewidth=0.8),
                    ha="center"
                   );

    ax.spines[["left", "top", "right", "bottom"]].set_visible(False);
    ax.spines[["bottom"]].set_position(("axes", 0.5));
    ax.yaxis.set_visible(False);
    ax.set_title("Timeline Pengalaman Ari", pad=10, loc="left", fontsize=50, fontweight="bold");
    ax.grid(False)
```
